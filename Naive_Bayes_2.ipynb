```python
import pandas as pd
import numpy as np
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from time import time
```

```python
#1622
#User 564
#Reading users file:
u_cols = ['user_id','age','sex','occupation','zip_code']
users = pd.read_csv('ml-100k/u.user', sep='|', \
                    names=u_cols,encoding='latin-1')

#Reading ratings file:
r_cols = ['user_id','movie_id','rating','unix_timestamp']
rating = pd.read_csv('ml-100k/u.data',sep='\t',names=r_cols,\
                     encoding='latin-1')

#Reading the movie file:
item_cols = ['movie_id','movie title','release date',
'video release date','IMDb URL','unknown','Action',
'Adventure','Animation','Childrens','Comedy','Crime',
'Documentary','Drama','Fantasy','Film-Noir','Horror',
'Musical','Mystery','Romance ','Sci-Fi','Thriller',
'War' ,'Western']

movies = pd.read_csv('ml-100k/u.item', sep='|',names=item_cols,\
                   encoding='latin-1')

df2  = rating.copy()

df2.head()

groups = df2.groupby("movie_id", as_index=False)
k = 0
columns = ['movie_id','Probability']
#print(users)
df_users = pd.DataFrame(columns=columns)
#print(df_user)
for group in groups:
    toy_stor = group
    toy_story = toy_stor[1].copy()
    #print((toy_story) )
    movie_num = toy_story['movie_id'].iloc[0]
    users_rating = users.copy()
    df_1_toy = users.copy()
    df_1_toy.occupation.unique()
    occupations = df_1_toy.occupation.unique()
    nums = list(range(len(occupations)))

    df_1_toy.sex.replace(['M', 'F'], [1, 0], inplace=True)
    df_1_toy.occupation.replace(occupations, nums, inplace=True)

    #df_1_toy.head()

    user_movie = pd.merge(df_1_toy,toy_story)
    user_movie = pd.merge(df_1_toy,toy_story,how='left')

    #print(df_1_toy.shape)

    #user_movie

    user_movie1 = user_movie.fillna(0, inplace=True)

    #user_movie1 = user_movie.drop(['unix_timestamp','rating','zip_code'],axis=1)
    #user_movie = user_movie.drop(['unix_timestamp','rating','zip_code'],axis=1)
    user_movie1 = user_movie.drop(['unix_timestamp','rating','zip_code','occupation'],axis=1)
    user_movie = user_movie.drop(['unix_timestamp','rating','zip_code','occupation'],axis=1)

    #print(list(user_movie))
    #print((user_movie1))

    train = user_movie1[:564].copy()#0-563, aka 1-564
    test = user_movie1[564:].copy()#564-942 aka 565-943

    #train.head()

    labels_train = user_movie['movie_id'][:564].copy()
    labels_test = user_movie['movie_id'][564:].copy()

    train = train.drop(['movie_id'],axis=1)
    test = test.drop(['movie_id'],axis=1)

    #print(train.head())
    #train.shape

    #labels_train.head()

    #print(labels_test.head())

    #train.info()

    train = train.reset_index(drop=True)

    #print(train.shape)
    #print(labels_train.shape)

    GNB = GaussianNB()

    t0 = time()
    GNB.fit(train, labels_train)
    #print("Training Time: ", time()-t0, "s.", sep='')

    t1 = time()
    nb_pred = GNB.predict(test)
    approx_pred = GNB.predict_proba(test)
    #print("Testing Time: ", time()-t1, "s.", sep='')

    #print("Accuracy: ", accuracy_score(labels_test, nb_pred), ".", sep='')
    #print("NotWatch\tWatch")
    #print(len(approx_pred[0]))
    #print(movie_num)
    #print("here")
    #print(approx_pred[1])
    #print(approx_pred[0])
    if(len(approx_pred[0]) == 2):
        df_temp = pd.DataFrame([[movie_num,approx_pred[:,1]]],columns=['movie_id','Probability'])
        #print(approx_pred[0][1])
    elif(approx_pred[0] == 1 and nb_pred[0] == 0):
        df_temp = pd.DataFrame([[movie_num,0]],columns=['movie_id','Probability'])
        #print(approx_pred[0])
    elif(approx_pred[0] == 1 and nb_pred[0] > 0):
        #print(approx_pred[0])
        df_temp = pd.DataFrame([[movie_num,1]],columns=['movie_id','Probability'])
    #df_temp = pd.DataFrame([[movie_num,approx_pred[:,1]]],columns=['movie_id','Probability'])
    df_users = df_users.append(df_temp)
    #print(approx_pred[0][1])
    #print(approx_pred[7])

    #print(nb_pred[0])
    #print(nb_pred[7])
    #print("Predicted\n")
    #print(nb_pred)
    #print(nb_pred.shape)
    #print("\n\nActual\n")
    #print(labels_test)
    #if(k==10):
     #   break
    k+=1
    #print("\n\n")
print(df_users)
```
